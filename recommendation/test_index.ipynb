```python
content_embeddings_with_movie_idx = pd.read_pickle("./pkl/autoencoder_embeddings-movie-index.pkl")
content_embeddings_with_movie_idx = pd.DataFrame(content_embeddings_with_movie_idx)
print(content_embeddings_with_movie_idx.shape)
content_embeddings_with_movie_idx.head()
```

```python
collaborative_embeddings_3 = pd.read_pickle("./pkl/nn_embeddings_3.pkl")
collaborative_embeddings_3 = pd.DataFrame(collaborative_embeddings_3)
print(collaborative_embeddings_3.shape)
collaborative_embeddings_3.head()
```

```python
def get_recs(movie_id, embeddings, df):
    sim_model = SimilarityPredictions(embeddings, similarity_metric="cosine")
    output = sim_model.predict_similar_items(seed_item=movie_id, n=20)
    return output
    similar_movies = pd.DataFrame(output)
    similar_movies.set_index('item_id', inplace=True)
    sim_df = pd.merge(df, similar_movies, left_index=True, right_index=True)
#     sim_df = pd.merge(df, similar_movies, left_on='movieId', right_index=True)
    sim_df.sort_values('similarity_score', ascending=False, inplace=True)
    #save recs locally
#     sim_df.head(20).to_csv(file_path, index=False, header=True)
    return sim_df.head(10)
```

```python
similar_items = get_recs(157, content_embeddings_with_movie_idx, movies_with_movie_index_rating)
```
